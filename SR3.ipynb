```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data= pd.read_csv("C:/Users/Sfundesihle/Desktop/test/population-and-demography.csv/population-and-demography.csv")
print(data.head())
```

```python
summary_stats = data.describe()
print(summary_stats)
```

```python
raws = 5
column_Name = 'Year'
data_column = data[column_Name]
data_raw = data_column.head(raws)

column = 'Population'
Name = data[column]
data = Name.head(raws)
plt.figure(figsize=(4,5))

```

```python
plt.plot(data_raw,data, marker='o',linestyle='-',color='red')
plt.title('Top 10 Years with Highest Number of popolation settlement')
plt.xlabel('Years')
plt.ylabel('population growth')
plt.xticks(rotation=45)
plt.show()

```

```python
plt.scatter(data_raw,data)
plt.title('Scatter Plot Of Years vs.Population')
plt.xlabel('Years')
plt.ylabel('Populations Growth Over The Years')
plt.show()
```

```python
plt.pie(data,labels=data_raw,autopct='%1.1f%%',startangle=140)
plt.title('Population Growth Over The Years') 
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot( x=data_raw, y=data, data=data)
plt.xlabel('Years')
plt.ylabel('Population Growth Over The Years')
plt.show()
```

```python
sns.jointplot(x=data_raw,y=data,data=data,kind='scatter')

```

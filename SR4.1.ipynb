```python
import pandas as pd

data=pd.read_csv("C:/Users/Sfundesihle/Desktop/test/population-and-demography.csv/population-and-demography.csv")
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
population_sorted = insertion_sort(data['Population'].tolist())
```

```python
data['Population_Sorted'] = population_sorted
```
